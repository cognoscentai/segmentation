# Wrong Object annotation 

15 (car), 20 (computer), 22 (drink), 27 (recorder), 47 (basket) 

```python
import pickle as pkl
```

The plot on the right is compounded with the area information since the number of votes are coupled with the pixels and not the tiles themselves . We need a way to figure out how many votes per tile. For each pixel in a tile, lookup in the mega_mask what its number of votes are. 

```python
wrong_objs = [15,20,22,27,47]
for wrong_obj in wrong_objs:
    img = pkl.load(open("30workers_rand0/obj{}/mega_mask.pkl".format(wrong_obj)))
    plt.figure(figsize=(10,4))
    plt.title("Heatmap of number of votes")
    plt.subplot(131)
    plt.imshow(img)
    plt.colorbar()
    plt.subplot(132)
    #     a = plt.hist(img.flatten()[np.where(img.flatten()!=0)],bins=30)
    tiles = pkl.load(open("30workers_rand0/obj{}/tiles.pkl".format(wrong_obj)))
    tvotes = []
    for t in tiles:
        if t!=set():
            tvotes.append(img[t.pop()])
        else:
            tvotes.append(0)
    plt.hist(tvotes,bins=30)

    plt.xlabel("Votes")
    plt.ylabel("Number of tiles")

    plt.subplot(133)
    tvotes = np.array(tvotes)
    canvas = np.zeros_like(img)
    plt.title("1-vote tiles")
    for tidx in np.where(tvotes==1)[0]:
        for pix in tiles[tidx]:
            canvas[pix]=1
    plt.imshow(canvas)

    plt.tight_layout()
```
